```python
# Import libraries
import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn.cluster import KMeans
import sklearn.metrics as sm
import pandas as pd
import numpy as np
from sklearn.cluster import AgglomerativeClustering
```

```python
# Load data
iris=datasets.load_iris()
```

```python
# Check class names
print(iris.target_names)
```

```python
# Split input and target features
X = pd.DataFrame(iris.data)
y = pd.DataFrame(iris.target)
```

```python
# Provide column name
X.columns=['Sepal_length','Sepal_width','Petal_length','Petal_width']
y.columns=['Target']
```

```python
# Define colormap for each class
colormap=np.array(['green','red','blue'])
```

```python
# Perform agglomerative clustering
Hcluster=AgglomerativeClustering(n_clusters=3)
Hcluster.fit(X)
predY=Hcluster.labels_
predY
```

```python
# Petal true labels
plt.scatter(X.Petal_length,X.Petal_width,c=colormap[y.Target])
plt.title('Class labels (Petal)')
plt.xlabel('Petal Length')
plt.ylabel('Petal Width')
```

```python
# Petal clusters
plt.scatter(X.Petal_length,X.Petal_width,c=colormap[predY])
plt.title('Hierarchical Classification (Petal)')
plt.xlabel('Petal Length')
plt.ylabel('Petal Width')
```

```python
# Sepal true labels
plt.scatter(X.Sepal_length,X.Sepal_width,c=colormap[y.Target])
plt.title('Class labels (Sepal)')
plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')
```

```python
# Sepal clusters
plt.scatter(X.Sepal_length,X.Sepal_width,c=colormap[predY])
plt.title('Hierarchical Classification (Sepal)')
plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')
```
